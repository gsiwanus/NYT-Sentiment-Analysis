```python
from textblob import TextBlob

#correct file name
file_name = 'NYT_20231201.txt'

try:
    with open(file_name, 'r', newline='', encoding="cp437", errors='ignore') as file:
        blob = TextBlob(file.read())
        print(f"Sentiment for the entire article:\n", blob.sentiment, "\n")
        

        # Display the sentiment for each sentence
        for i, sentence in enumerate(blob.sentences):
            print(f"Sentence {i+1}:\n", sentence.sentiment, "\n")
            
except FileNotFoundError:
    print(f"The file '{file_name}' was not found.")
except Exception as exception:
    print(f"An error occurred: {exception}")
```
